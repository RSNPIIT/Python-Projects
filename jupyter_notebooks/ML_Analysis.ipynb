```python
import pandas as pd

# URLs converted to direct download links
train_url = f'https://drive.google.com/uc?id=1ZqdDPJm3qF6m3nH6juMAnJoYtothbpUK'
test_url = f'https://drive.google.com/uc?id=1jrk-d7NQvhLBAJyYNEJBuRB33Wv9jVuL'

train_df = pd.read_csv(train_url)
test_df = pd.read_csv(test_url)

print("Files loaded successfully!")
print(f"Train size: {len(train_df)}")
print(f"Test size: {len(test_df)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import warnings

# iii. Ignore warnings
warnings.filterwarnings("ignore")

# vii. & ix. Separate dependent and independent variables
# Assuming the last column is the target (y) and everything else is X
X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]
X_test = test_df.iloc[:, :-1]
y_test = test_df.iloc[:, -1]

# x. Run LogisticRegression with random state = 17
log_reg = LogisticRegression(random_state=17)

# xi. Fit the model on X_train and y_train
log_reg.fit(X_train, y_train)

# xii. Predict on training data
y_pred_train = log_reg.predict(X_train)

# xiii. Predict on test data
y_pred_test = log_reg.predict(X_test)

# xiv. Print classification report for training and test (rounded to 4 decimals)
print("--- Training Data Classification Report ---")
print(classification_report(y_train, y_pred_train, digits=4))

print("\n--- Test Data Classification Report ---")
print(classification_report(y_test, y_pred_test, digits=4))

# xv. Create the confusion matrix for test data
print("\n--- Test Data Confusion Matrix ---")
cm = confusion_matrix(y_test, y_pred_test)
print(cm)
```

```python
# Extracting counts from your test dataset results
total_test = 36415
not_accepted_test = 28350
accepted_test = 8065

# Calculating proportions as percentages
prop_not_accepted_test = (not_accepted_test / total_test) * 100
prop_accepted_test = (accepted_test / total_test) * 100

print(f"Proportion Not Accepted (Test): {prop_not_accepted_test:.4f}")
print(f"Proportion Accepted (Test): {prop_accepted_test:.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score

# 1. Calculate Training Accuracy
# We multiply by 100 to get the percentage value
train_accuracy = accuracy_score(y_train, y_pred_train) * 100

# 2. Calculate Training Precision for Class 0 (Not accept the loan)
# We set pos_label=0 because the question asks specifically about customers who will NOT accept.
train_precision_0 = precision_score(y_train, y_pred_train, pos_label=0) * 100

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Training Precision (Class 0): {train_precision_0:.4f}")
```

```python
from sklearn.metrics import recall_score, confusion_matrix

# 1. Training data recall for "Accepted" (Class 1)
# We use pos_label=1 because the aim is to find customers who DID accept.
train_recall_1 = recall_score(y_train, y_pred_train, pos_label=1) * 100

# 2. Extract Type 1 Error (False Positives) from the test data confusion matrix
# .ravel() flattens the matrix into: True Negative, False Positive, False Negative, True Positive
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()

print(f"1. Training Recall (Class 1) in %: {train_recall_1:.4f}")
print(f"2. Number of False Positives (Type 1 Error): {fp}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. Extract Type 2 Error (False Negatives) from the test data confusion matrix
# .ravel() flattens the matrix into: tn, fp, fn, tp
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()

# 2. Calculate Test Accuracy
# Multiply by 100 for percentage and round to 2 decimals
test_accuracy = round(accuracy_score(y_test, y_pred_test) * 100, 2)

print(f"Number of False Negatives (Type 2 Error): {fn}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
from sklearn.metrics import precision_score, recall_score

# 1. Test Precision for Class 0 (Not accept the loan)
test_precision_0 = precision_score(y_test, y_pred_test, pos_label=0) * 100

# 2. Test Recall for Class 1 (Accept the loan)
test_recall_1 = recall_score(y_test, y_pred_test, pos_label=1) * 100

print(f"Test Precision (Class 0): {round(test_precision_0, 2)}")
print(f"Test Recall (Class 1): {round(test_recall_1, 2)}")
```
